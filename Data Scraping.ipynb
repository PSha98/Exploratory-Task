```python
import requests
import pandas as pd
df_merged=pd.DataFrame()
url = "https://www.startupwala.com/list-of-registered-companies-in-india-P"
for e in [str(x) for x in range(1,11)]:
    df_merged = df_merged.append(pd.read_html(requests.get(url+e).text)[0])
df_merged.reset_index(drop=True, inplace=True)
df_merged
```
