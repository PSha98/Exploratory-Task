```python
import mysql.connector
import requests
import pandas as pd
```

```python
mydb = mysql.connector.connect(
    host = "localhost",
    user = "root",
    password = "",
    database = "company"
)

cur = mydb.cursor()

cur.execute("SELECT * FROM company")
for row in cur.fetchall():
    print(row)
```

```python
df=pd.DataFrame()

for e in [str(x) for x in range(1,11)]:
    df = df.append(pd.read_html(requests.get("https://www.startupwala.com/list-of-registered-companies-in-india-P"+e).text)[0])

    df.reset_index(drop=True, inplace=True)
df
```

```python
b,c,d,e=df['COMPANY NAME'].values,df['CIN'].values,df['ROC NAME'].values,df['STATUS'].values

for i in range(0,len(df['COMPANY NAME'].values)):
    sql = "INSERT INTO company (companyname,cin,city,status) VALUES (%s,%s,%s,%s)"
    val = (b[i],c[i],d[i],e[i])
    cur.execute(sql, val)
    mydb.commit()
print(cur.rowcount, "record inserted.")

cur.execute("SELECT * FROM company")
for row in cur.fetchall():
    print(row)
```
